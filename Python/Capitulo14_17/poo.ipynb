# Classe copm método e propriedade 

```python
class Horario:
    """Representa um horário do dia"""
    def __init__(self, hora=0, minuto=0, segundo=0):
        self.hora = hora
        self.minuto = minuto
        self.segundo = segundo
    
    def __str__(self):
        return f'{self.hora:02d}:{self.minuto:02d}:{self.segundo:02d}'
    
    def horario_to_inteiro(self):
        minutos = self.hora  * 60 + self.minuto
        segundos = minutos *  60  + self.segundo
        return segundos
    
    def inteiro_to_horario(segundos):
        minuto, segundo = divmod(segundos, 60)
        hora, minuto = divmod(minuto,  60)
        return Horario(hora, minuto, segundo)

    def __add__(self, outro):
        segundos = self.horario_to_inteiro() + outro.horario_to_inteiro()
        return Horario.inteiro_to_horario(segundos)

    def descreverHorario(self):
        mensagem = f'Este Objeto possui a hora{self.hora:02d}, '
        mensagem = mensagem + f'o minuto é {self.minuto:02d}' 
        mensagem = mensagem + f'o segundo é {self.segundo:02d}. '
        return mensagem

```

```python
natacao = Horario(20,15,0)
```

```python
print(natacao)
```

```python
natacao.descreverHorario()
```

# Polimorfismo 

```python
class Operacao:
    def __init__(self, x, y):
        self.x = x 
        self.y = y

    def calcular(self):
        pass 

    
```

```python
class Soma(Operacao):
    def calcular(self):
        return self.x + self.y
```

```python
total = Soma(10,20)
total.calcular()
```

```python
class Multiplicacao(Operacao):
    def calcular(self):
        return self.x * self.y
```

```python
total = Multiplicacao(240,3)
total.calcular()
```
